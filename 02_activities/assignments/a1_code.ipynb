```python
#adding titles etc.

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title("total growth over time")
ax.set_ylabel("total growth")
ax.set_xlabel("years")
fig.tight_layout() #keeps everything together, not always necessary
```

```python
#dictionary for font info

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title("total growth over time", fontdict = font1)
ax.set_ylabel("total growth", fontdict= font2)
ax.set_xlabel("years", fontdict=font2)
fig.tight_layout()
```

```python
#change positions of labels

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title("total growth over time", fontdict = font1, loc='left')
ax.set_ylabel("total growth", fontdict= font2)
ax.set_xlabel("years", fontdict=font2)
fig.tight_layout()
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1,y2)
#line plot with two lines
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'person_1')
ax.plot(x,y2, label = "person_2")
ax.legend(loc = "lower right")
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'person_1')
ax.plot(x,y2, label = "person_2")
ax.legend(loc = 'lower right', 
          frameon = True, fontsize = 12, shadow = True)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'person_1')
ax.plot(x,y2, label = "person_2")
ax.legend(loc = 'upper left', bbox_to_anchor = (1,1))
#loc acts as an anchor point
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = 'person_1')
ax.scatter(x,y2, label = "person_2")
ax.legend(loc = 'lower right')

ax.text(10,95, "this value is important", ha = 'center', color = 'red', size = 20)
#position based on data coords
```

```python
#moving text around based on the axes/data/figure

fig, ax = plt.subplots ()
ax.axis([0,10,0,10])
ax.text(1,5, "data(1,5)", transform = ax.transData)
ax.text(.5, .1, 'axes(.5, .1)', transform = ax.transAxes)
ax.text(.2, .2, 'figure (.2, .2)', transform = fig.transFigure)
```

```python
#adding arrows/flags

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'person 1')
ax.scatter(x,y2, label = 'person 2')
ax.legend(loc='lower right')

ax.annotate('This is important!', 
            xy=(10,95), xytext=(20,94),
            arrowprops=dict(arrowstyle = 'wedge', color = 'hotpink')) #arrow shape and color
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ="person 1")
ax.scatter(x,y2,label = "person2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
#set x and y axis to be blank
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
#set number of x axis tick marks
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "person 1")
ax.scatter(x, y2, label = "person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(10))
#sets the interval between the x axis labels
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "person 1")
ax.scatter(x,y2,label = "person 2")
ax.legend(loc='lower right')

plt.xticks(rotation = 45, ha='right')
#horizontal x axis labels
```

```python
#seaborn stuff
#deals w/ semantic mapping


#data
tips = sns.load_dataset("tips")
print(tips)

```

```python
#line plot

sns.lineplot(data = tips, x = 'total_bill', y = "tip")
```

```python
#line plot
sns.set_style('dark')
sns.lineplot(data = tips, x = 'total_bill', y = "tip")
```

```python
#titles/labels 

tipgraph = sns.lineplot(data = tips, x = 'total_bill', y = "tip")
tipgraph.set(title = "tips vs. bill total",
             xlabel = "total bill in dollars",
             ylabel = "tip amount in dollars")

tipgraph
```

```python
#titles/labels 
fig = plt.subplots(figsize = (10,4)) #manual size

tipgraph = sns.lineplot(data = tips, x = 'total_bill', y = "tip",
                        color = "pink", linestyle = "--", marker = 'o') #can also add linewidth, etc.

tipgraph.set(title = "tips vs. bill total",
             xlabel = "total bill in dollars",
             ylabel = "tip amount in dollars")

tipgraph
```

```python
#setting colour groupings

tipgraph = sns.scatterplot(data = tips, x= "total_bill", y="tip", 
                           style = "time", hue = "day", palette= ['purple', 'pink', 'skyblue', 'yellow'])

#style sets the marker grouping (circle and cross)
```

```python
#pairplots for comparing variables

sns.pairplot(data = tips, hue = "day")
```

```python
#relplot for multiple plots looking within a grouping varialbe

dayplot = sns.relplot(data = tips, x = 'total_bill', y = 'tip',
                     hue = 'sex', col = 'day', kind = 'scatter',
                     col_wrap = 2)
```

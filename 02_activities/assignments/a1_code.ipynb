```python
#libraries for simple graph

import scipy
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)
#values from 0-100, 50 values

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y)
```

```python


fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y)

#for horizontal
#fig, ax = plt.subplots(figsize = (5,3))
#ax.barh(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.hist(y)
```

```python
#adding titles etc.

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title("total growth over time")
ax.set_ylabel("total growth")
ax.set_xlabel("years")
fig.tight_layout() #keeps everything together, not always necessary
```

```python
#dictionary for font info

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title("total growth over time", fontdict = font1)
ax.set_ylabel("total growth", fontdict= font2)
ax.set_xlabel("years", fontdict=font2)
fig.tight_layout()
```

```python
#change positions of labels

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title("total growth over time", fontdict = font1, loc='left')
ax.set_ylabel("total growth", fontdict= font2)
ax.set_xlabel("years", fontdict=font2)
fig.tight_layout()
```

```python
#modify data points

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
        marker = '*',
        color = "indigo")

```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
        marker = '*',
        color = "indigo",
        linestyle = '--',
        linewidth = 2)

```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 10,
        color = "indigo", #can also use hex codes
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = 'blue',
        markerfacecolor = 'green')
#ax.grid() i hate grid lines i will never use them
#ax.grid(axis = 'y', color = 'black', linewidth = .5, linestyle = '-.')

```

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.pie(y, labels=x, autopct='%1.1f%%')
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1,y2)
#line plot with two lines
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'person_1')
ax.plot(x,y2, label = "person_2")
ax.legend(loc = "lower right")
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'person_1')
ax.plot(x,y2, label = "person_2")
ax.legend(loc = 'lower right', 
          frameon = True, fontsize = 12, shadow = True)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'person_1')
ax.plot(x,y2, label = "person_2")
ax.legend(loc = 'upper left', bbox_to_anchor = (1,1))
#loc acts as an anchor point
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = 'person_1')
ax.scatter(x,y2, label = "person_2")
ax.legend(loc = 'lower right')

ax.text(10,95, "this value is important", ha = 'center', color = 'red', size = 20)
#position based on data coords
```

```python
#moving text around

fig, ax = plt.subplots ()
ax.axis([0,10,0,10])
ax.text(1,5, "data(1,5)", transform = ax.transData)
ax.text(.5, .1, 'axes(.5, .1)', transform = ax.transAxes)
ax.text(.2, .2, 'figure (.2, .2)', transform = fig.transFigure)
```
